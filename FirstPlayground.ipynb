```julia
# bring in modules
include.(("./lib/demand.jl","./lib/generator.jl","./lib/market.jl","./lib/plotround.jl"))
using .Generator: isless # this allows sort to function
using .Demand: isless # this allows sort to function

# config
n_market_periods = 26
verbose = false

# import bids
imported_generators = Generator.importFromFile("generator_test.yaml")
verbose && println(imported_generators)

imported_demands = Demand.importFromFile("demand_test.yaml")
verbose && println(imported_demands)

# clear market
marketOutcomes = Market.Clear(n_market_periods, imported_generators, imported_demands)

verbose && println(marketOutcomes)

# plot

for (iter, marketOutcome) in enumerate(marketOutcomes)
    marketPlot = PlotRound.PlotMarketOutcome(iter, imported_generators, imported_demands, marketOutcome)
    display(marketPlot)
end
```
